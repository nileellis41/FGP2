```python
import yfinance as yf
import pandas as pd

def get_last_earnings_for_stocks(tickers):
    all_last_earnings = []

    for ticker in tickers:
        try:
            # Get ticker data
            stock = yf.Ticker(ticker)
            
            # Get earnings history
            earnings_history = stock.earnings_history
            
            # Check if earnings_history is valid
            if earnings_history is None or earnings_history.empty:
                print(f"No earnings history available for ticker: {ticker}")
                continue
            
            # Get the last row and add ticker as a column
            last_row = earnings_history.iloc[[-1]].copy()
            last_row['Ticker'] = ticker
            last_row.index = [f"Earnings for {ticker} on {last_row.index[0]}"]
            
            all_last_earnings.append(last_row)
        
        except Exception as e:
            print(f"An error occurred for ticker {ticker}: {e}")

    # Combine all rows into a single DataFrame
    combined_df = pd.concat(all_last_earnings) if all_last_earnings else pd.DataFrame()
    return combined_df

# Example usage with 4 stocks
tickers = ["ASC", "AXR", "CHCI", "EGY", "GAMB", "IDR", "MDXG", "OPXS", "RDVT"]
last_earnings_df = get_last_earnings_for_stocks(tickers)

# Display the result
if not last_earnings_df.empty:
    print(last_earnings_df)
```
