```python
import yfinance as yf
import pandas as pd

def get_fundamentals(symbol):
    # Fetch the stock data
    stock = yf.Ticker(symbol)

    # Extracting the fundamentals
    info = stock.info

    fundamentals = {
        'Symbol': symbol,
        'Name': info.get('longName', 'N/A'),
        'Exchange': info.get('exchange', 'N/A'),
        'Industry': info.get('industry', 'N/A'),
        'Sector': info.get('sector', 'N/A'),
        'CEO': info.get('ceo', 'N/A'),
        'Employees': info.get('fullTimeEmployees', 'N/A'),
        'Headquarters': info.get('address1', 'N/A') + ', ' + info.get('city', 'N/A') + ', ' + info.get('state', 'N/A'),
        'Description': info.get('longBusinessSummary', 'N/A'),
        'Market Cap': info.get('marketCap', 'N/A'),
        'Dividend Rate': info.get('dividendRate', 'N/A'),
        'Dividend Yield': info.get('dividendYield', 'N/A'),
        'Earnings per Share': info.get('epsTrailingTwelveMonths', 'N/A'),
        'Price to Earnings Ratio': info.get('trailingPE', 'N/A'),
        'Price to Book Ratio': info.get('priceToBook', 'N/A'),
        'Price to Sales Ratio': info.get('priceToSalesTrailing12Months', 'N/A'),
    }

    return pd.DataFrame([fundamentals])

# Example usage
symbol = 'AMD'
fundamentals_df = get_fundamentals(symbol)

import ace_tools_open as tools; tools.display_dataframe_to_user(name="AMD Fundamentals", dataframe=fundamentals_df)

print(fundamentals_df)
```
