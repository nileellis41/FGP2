```python
import yfinance as yf
import pandas as pd

# Load the CSV file containing stock holdings
csv_file_path = 'IWC_holdings.csv'  # Adjust the path if necessary
iwc_holdings_df = pd.read_csv(csv_file_path)

# Extract the tickers
tickers_list = iwc_holdings_df['Ticker'].tolist()

# Fetch current price data from Yahoo Finance
current_price_data = yf.download(tickers_list, period='1d')['Adj Close']

# Transpose the current price data
transposed_data = current_price_data.T

# Filter the transposed data for prices under 30
filtered_transposed_data_under_30 = transposed_data[transposed_data.iloc[:, 0] < 30]

# Drop any rows with NaN values if needed
filtered_transposed_data_under_30 = filtered_transposed_data_under_30.dropna()

# Export the filtered DataFrame to CSV
filtered_transposed_data_under_30.to_csv('filtered_stocks_under_30.csv', index=True)

# Optionally display the filtered DataFrame
print(filtered_transposed_data_under_30)
```
